```python
with open("data/atex.md", 'r', encoding='utf-8') as f:
    content= f.read()

print("첫번째 문서 로드")
print(f"문서 길이: {len(content)} 문자")
print(f"[Document] : \n\n{content[:100]}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 20,
    separators  = ["\n#", "\n\n", "\n\n#"]
)

```

```python
docs = text_splitter.create_documents([content])

print(f"청킹 사이즈 : {len(docs)}")
print(f"청크 타입 : {type(docs[0])}")

for doc in docs:
    print(f"[page_content] : \n\n {doc.page_content}")
    print(f"[metadata] : \n\n {doc.metadata}")
    print("-"*100)
```

1. mixed (기본값, 가장 많이 씀)
복합어 전체 + 분해된 단어들을 모두 토큰으로 포함

예시: "학생복지관" → "학생복지관", "학생", "복지", "관"

장점:

복합어 전체와 분해된 단어가 모두 검색에 걸림

대부분 검색 정확도 높임 (권장)

2. none
복합어 전체만 토큰화, 분해 안 함

예시: "학생복지관" → "학생복지관"

장점:

분해를 원치 않는, 예를 들어 고유명사 위주 데이터에 적합

"학생"만 검색해도 "학생복지관"이 검색되지 않음

3. discard
분해된 단어들만 토큰화, 복합어 전체는 버림

예시: "학생복지관" → "학생", "복지", "관"

장점:

복합어로 검색되는 걸 막고, 단어 조각 위주로 검색

"복지"만 검색해도 "학생복지관"이 결과에 포함



```python
nori_mapping = {
    "settings": {
        "analysis": {
            "tokenizer": {
                "nori_tokenizer": {
                    "type": "nori_tokenizer",
                    "decompound_mode": "discard"
                }
            },
            "filter": {
                "korean_stop": {
                    "type": "stop",
                    "stopwords": ["_korean_"]
                }
            },
            "analyzer": {
                "custom_korean": {
                    "type": "custom",
                    "tokenizer": "nori_tokenizer",
                    "filter": [
                        "lowercase",
                        "korean_stop"
                    ]
                }
            }
        }
    },
    "mappings": {
        "properties": {
            "content": {
                "type": "text",
                "analyzer": "custom_korean"
            },
            "chunk_id": {"type": "integer"}
        }
    }
}

test_index = "bm25_nori_test"

if es.indices.exists(index=test_index):
    es.indices.delete(index=test_index)
    print("기존 테스트 인덱스 삭제")

es.indices.create(index=test_index, body = nori_mapping)
```

```python
# decompose mixed test

test_text = "학생복지관"

response = es.indices.analyze(
    index=test_index,
    body = {"analyzer" : "custom_korean" , "text" : test_text}
)
```

```python
tokens = [token['token'] for token in response['tokens']]
print(f"✅ 분석 테스트 성공: '{test_text}' → {tokens}")

print(response)
```

k1

(기본값 1.2)

개별 단어가 문서 내에 여러 번 나올 때, 빈도수 증가에 따라 점수를 얼마나 많이 올릴지 결정

값이 커질수록: 같은 단어가 반복되면 점수 급격히 증가

값이 작을수록: 단어 빈도가 점수에 미치는 영향이 적어짐

"단어 등장 횟수의 중요도" (TF에 곱해짐)

b

(기본값 0.75)

문서 길이 보정 정도. 문서 길이가 평균에서 벗어날수록 점수에 미치는 영향

1에 가까울수록: 길이가 긴 문서는 점수가 깎임, 짧은 문서는 보정



```python
nori_mapping = {
    "settings": {
        "index" :{
            "similarity" : {
                "korean_bm25" : {
                    "type" : "BM25",
                    "k1" : 1.2,
                    "b" : 0.75
                }
            }
        },
        "analysis": {
            "tokenizer": {
                "nori_tokenizer": {
                    "type": "nori_tokenizer",
                    "decompound_mode": "discard"
                }
            },
            "filter": {
                "korean_stop": {
                    "type": "stop",
                    "stopwords": ["_korean_"]
                }
            },
            "analyzer": {
                "custom_korean": {
                    "type": "custom",
                    "tokenizer": "nori_tokenizer",
                    "filter": [
                        "lowercase",
                        "korean_stop"
                    ]
                }
            }
        }
    },
    "mappings": {
        "properties": {
            "content": {
                "type": "text",
                "similarity" : "korean_bm25",
                "analyzer": "custom_korean"
            },
            "chunk_id": {"type": "integer"}
        }
    }
}

test_index = "bm25_nori_test"

if es.indices.exists(index=test_index):
    es.indices.delete(index=test_index)
    print("기존 테스트 인덱스 삭제")

es.indices.create(index=test_index, body = nori_mapping)
```

| ------ | --- | ------------- | ---------- |
| **k1** | 0   | 무제한(권장: 3 이하) | 1.0 \~ 2.0 |
| **b**  | 0   | 1             | 0.5 \~ 0.8 |

```python
def search_with_analyzer(query, top_k = 5):
    search_body = {
        "size" : top_k,
        "query" : {
            "match" : {
                "content" : {
                    "query" : query,
                    "operator" : "or"
                }
            }
        },
        "_source" : ["content", "chunk_id", "id", "source"]
    }

    response = es.search(index = test_index, body = search_body)

    results = []

    for hit in response['hits']['hits']:
        results.append({
            "chunk_id" : hit['_source']['chunk_id'],
            "score" : hit['_score'],
            'content' : hit['_source']['content'],
            'source' : hit['_source'].get('source', 'unknown')
        })

        print("[HIT] \n\n")
        print(f"{hit}")

    return results
```

```python
query = "서창녕은 누구야?"

results = search_with_analyzer(query, top_k = 10)

for i , result in enumerate(results, 1):
    print(f"   [{i}] Chunk {result['chunk_id']} - 점수: {result['score']:.4f}")
    print(f"       내용: {result['content']}")
```

새로운 문서 추가가

1. operator: "or"의 의미
match 쿼리에서

operator: "or":
쿼리 문자열이 "학생 복지관"이라면
"학생" 또는 "복지관"이 들어간 문서 모두 검색

둘 중 하나만 있어도 OK

operator: "and":
"학생" 그리고 "복지관" 모두 포함된 문서만 검색

즉, "or"는 가장 많이 쓰는 디폴트(넓은 범위 검색)
"and"는 좁은 검색(정확한 문서만) 원할 때 사용

2. 자주 쓰는 match 쿼리의 추가 옵션들
"analyzer" : 사용할 분석기 지정(보통 인덱스에 매핑된 분석기 자동 적용됨)

"minimum_should_match" : 여러 키워드 중 최소 몇 개 이상 포함된 문서만 검색

json
코드 복사
{
  "match": {
    "content": {
      "query": "학생 복지관 안내",
      "minimum_should_match": "2"
    }
  }
}
"fuzziness" : 오타/유사어 허용(예: "fuzziness": "AUTO")

3. 검색 쿼리에서 쓸 수 있는 주요 키워드
"sort" : 점수 이외 다른 기준 정렬(예: 날짜순)

json
코드 복사
"sort": [
  {"date": "desc"}
]
"highlight" : 검색 키워드 하이라이트

json
코드 복사
"highlight": {
  "fields": {
    "content": {}
  }
}
"from"/"size" : 페이지네이션 (몇 번째부터, 몇 개)

"explain": true : 점수 계산 로직 설명 반환(디버깅용)

```python
from elasticsearch import helpers
```

```python
with open("data/seoul_rebuild_project.md", 'r', encoding='utf-8') as f:
    bulk_content = f.read()

bulk_docs = text_splitter.create_documents([bulk_content])

print(f"새로운 문서 길이: {len(bulk_content)} 문자")
print(f"새로운 문서 청크 개수: {len(bulk_docs)}")
```

```python
actions = []

start_id = len(docs) + len(new_docs)

for i, doc in enumerate(bulk_docs):
    action = {
        "_index" : test_index,
        "_id" : start_id + i,
        "_source" : {
            "content" : doc.page_content,
            "chunk_id" : start_id + i,
            "source" : "seoul_rebuild_project.md"
        }
    }

    actions.append(action)

helpers.bulk(es, actions)

es.indices.refresh(index= test_index)
```

```python
query = "서울시의회가 진행하는 프로젝트 이름름"

results = search_with_analyzer(query, top_k = 10)

for i, result in enumerate(results, 1):
    print(f"   [{i}] Chunk {result['chunk_id']} - 점수: {result['score']:.4f}")
    print(f"       내용: {result['content'][:100]}...")
    print(f"       출처: {result['source']}")
```
